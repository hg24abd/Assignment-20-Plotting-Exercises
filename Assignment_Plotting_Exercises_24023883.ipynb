Problem 1: Title: Exploring Random Data: Visualising Mean and Median with a Histogram

```python
# Generate beta-distributed random data
np.random.seed(42) # For Reproducibility
data = np.random.beta(a=2, b=5, size=300)
print(data)
```

```python
# Compute Statistics
mean = np.mean(data)
median = np.median(data)
```

```python
# Plot Histogram with Creative Enhancements

# Create the histogram
plt.figure(figsize=(10, 6))
counts, bins, patches = plt.hist(data, bins = 15, color = 'lightblue', alpha = 0.7, edgecolor = 'black', label = 'Data')

# Add shaded area under the histogram
for patch in patches:
    patch.set_facecolor('skyblue')
    patch.set_alpha(0.5)

# Mark the mean and median with lines
plt.axvline(mean, color = 'darkorange', linestyle = '--', linewidth = 2, label = f'Mean: {mean:.2f}')
plt.axvline(median, color = 'green', linestyle = '--', linewidth = 2, label = f'Median: {median:.2f}')

# Add annotations for mean and median
plt.text(mean, max(counts) * 0.8, f'Mean: {mean:.2f}', color = 'darkorange', fontsize = 12, ha = 'center')
plt.text(median, max(counts) * 0.7, f'Median: {median:.2f}', color = 'green', fontsize = 12, ha = 'center')

# Improve the plot aesthetics
plt.title('Histogram of Beta-Distributed Data', fontsize = 18, fontweight = 'bold')
plt.xlabel('Value', fontsize = 14)
plt.ylabel('Frequency', fontsize = 14)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.legend(fontsize = 12)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)

# Final layout adjustments
plt.tight_layout()
plt.show()
```

Problem 2: Title: Comprehensive Visual Analysis of Dataset from Parquet File

```python
# Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Load the data
file_url = "https://star.herts.ac.uk/~kuhn/DHV/exercises_problem2.parquet"
data = pd.read_parquet(file_url)
```

```python
# Step 2: Explore the dataset
# Display basic information about the dataset to understand its structure
print(data.head())
print(data.info())
print(data.describe())
print(data.isnull().sum())
```

```python
# Step 3: Pair Plot
# Purpose: To explore pairwise relationships between numerical variables.
sns.set_theme(style="whitegrid")  # Set a consistent theme for all plots
sns.pairplot(data, diag_kind="kde", height=2)
plt.suptitle("Pairwise Relationships (Numerical Variables)", y=1.02)
plt.show()
```

```python
# Step 4: Correlation Heatmap
# Purpose: To show the strength and direction of relationships between numerical variables.
plt.figure(figsize=(10, 8))
correlation = data.corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Step 5: Distribution Plots (Histograms with KDE)
# Purpose: To visualise the distribution of numerical data and detect skewness or multimodality.
for col in data.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(data[col], kde=True, bins=30, color='skyblue')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Step 6: Box Plots for Outliers
# Purpose: To identify outliers and compare distributions across multiple categories.
for col in data.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=data[col], color="lightcoral")
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()
```

```python
# Step 7: Count Plot for Categorical Variables
# Purpose: To display the frequency of categories in each categorical column.
for col in data.select_dtypes(include=['object', 'category']).columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=data[col], palette="Set2")
    plt.title(f"Count of {col}")
    plt.xticks(rotation=45)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
# Step 8: Violin Plots (Combining Categorical and Numerical Data)
# Purpose: To visualise the distribution of a numerical variable for each category.
for cat_col in data.select_dtypes(include=['object', 'category']).columns:
    for num_col in data.select_dtypes(include=['float64', 'int64']).columns:
        plt.figure(figsize=(8, 6))
        sns.violinplot(x=cat_col, y=num_col, data=data, palette="muted")
        plt.title(f"Distribution of {num_col} by {cat_col}")
        plt.xlabel(cat_col)
        plt.ylabel(num_col)
        plt.xticks(rotation=45)
        plt.show()
```

```python
# Step 9: Multivariable Analysis (Scatter Plot with Hue)
# Purpose: To explore how a numerical variable is influenced by a categorical variable.
# Example: Scatter plot with colour-coded categories
for cat_col in data.select_dtypes(include=['object', 'category']).columns:
    for num_col1 in data.select_dtypes(include=['float64', 'int64']).columns:
        for num_col2 in data.select_dtypes(include=['float64', 'int64']).columns:
            if num_col1 != num_col2:  # Avoid plotting a variable against itself
                plt.figure(figsize=(8, 6))
                sns.scatterplot(x=num_col1, y=num_col2, hue=cat_col, data=data, palette="viridis")
                plt.title(f"Scatter Plot of {num_col1} vs {num_col2} by {cat_col}")
                plt.xlabel(num_col1)
                plt.ylabel(num_col2)
                plt.legend(title=cat_col)
                plt.show()
```
